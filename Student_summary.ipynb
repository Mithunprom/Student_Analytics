```julia
names(df)
#df."Avg Transfer Units"
#A=df."Avg. Pass Units (UA)",df."Avg. UA Units Taken Not Applied",df."Average Total Non-UA Units"
#convert(Matrix, df[:,1:3])
A=df[:,[1,9,6,5,12]]
```

```julia
#A1=sortrows(A, by= x=x[1])
A=convert(Matrix, df[:,[1,9,6,5,12]])
A=sortslices(A,dims=1,by=x->x[2]+x[3]+x[4],rev=true)
A1=A[:,2:4]
A1=convert(Array{Float64,2}, A1)
A2=A[:,1]
A3=A[:,5]
```

```julia
plotly()
B1=groupedbar(A1, bar_position = :stack, 
    bar_width=0.7,
    xticks=(1:15,A2),
    xrotation=55,
    fillalpha=0.25,
    legend=:topright,
    labels=["Average Total Non-UA Units" "Avg. UA Units Taken Not Applied" "Avg. Pass Units (UA)"],
)
B2=plot!(1:15, A3, linetype=:steppost,linestyle=:dot,labels="Minimum Required Credits",
    arrow = :arrow,
    linealpha = 0.5,
    linewidth = 4,
    linecolor = :red)
ylims!((0,200))
fnt = Plots.font("Aria", .001)
default(titlefont=fnt, guidefont=fnt, tickfont=fnt, legendfont=fnt)
plot!(size=(800,1100))

display("text/html",B2)
#display("text/html",B1)
#[B1, B2]
```

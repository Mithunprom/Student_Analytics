```julia
plotly()
B1=groupedbar(A1, bar_position = :stack, 
    bar_width=.7,
    xticks=(1:15,A2),
    xrotation=55,
    fillalpha=1,
    legend=:topright,
    labels=["Avg. Non-UA Units" "Avg. UA Units-Not Applied" "Avg. UA Units-Applied"],hover=A1
)
[annotate!(i, a[i], text(ab[i],10), :color) for i in 1:15]
#[annotate!(i,40, text(A1[i,3],8), :color) for i in 1:15]
#[annotate!(i,A1[i,3]+A1[i,2]-3, text(A1[i,2],8), :color) for i in 1:15]
#[annotate!(i,A1[i,3]+A1[i,2]+A1[i,1]-3, text(A1[i,1],8), :color) for i in 1:15]

B2=plot!(.5:1:16, A3, linetype=:steppost,linestyle=:dot,labels="Minimum Required Credits",
    arrow = :arrow,
    linealpha = 0.5,
    linewidth = 4,
    linecolor = :red,
    hover=A3)

ylims!((0,200))
#fnt=Plots.font("Aria",.001)
default(titlefont=fnt, guidefont=fnt, tickfont=fnt, legendfont=fnt)
plot!(size=(800,900))
display("text/html",B2)
```

```julia
df1=CSV.read("Units_Student_summary.csv")
```

```julia
df1.Units_taken_but_not_complted=df1[:,"TotalUnitsTakentowardsDegree"]-df1[:,"TotalUnitsPassedtowardsDegre"];
df1 = df1[completecases(df1), :];
```

```julia
groupedbar(df1.College, df1.GPADegree, group = df1.IPEDSEthnicity, xlabel = "Groups", ylabel = "GPA", bar_width = 0.67,
        lw = 0, framestyle = :box,
        title = "Scores by group and category",xrotation=55,size=(1000,1000)) 
ylims!((0,7))
```

```julia
#df1 |> @filter(_.College âˆˆ ["Colleges of Letters Arts & Sci"]) |> DataFrame
clg_architecture=df1[findall(in(["Colleges of Letters Arts & Sci"]), df1.College), :]
clg_humanities=df1[findall(in(["College of Humanities"]), df1.College), :]
clg_Agre_life_science=df1[findall(in(["College of Agric and Life Sci"]), df1.College), :]
clg_applied_science=df1[findall(in(["College of Applied Sci & Tech"]), df1.College), :]
clg_pland_landscape=df1[findall(in(["Col Arch Plan & Landscape Arch"]), df1.College), :]
clg_education=df1[findall(in(["College of Education"]), df1.College), :]
clg_engineering=df1[findall(in(["College of Engineering"]), df1.College), :]
clg_optical=df1[findall(in(["James C Wyant Coll Optical Sci"]), df1.College), :]
clg_medicine=df1[findall(in(["College of Medicine - Tucson"]), df1.College), :]
clg_nurshing=df1[findall(in(["College of Nursing"]), df1.College), :]
clg_science=df1[findall(in(["College of Science"]), df1.College), :]
clg_social_bhvr=df1[findall(in(["College of Social & Behav Sci"]), df1.College), :]
clg_eller=df1[findall(in(["Eller College of Management"]), df1.College), :]
clg_fine_arts=df1[findall(in(["College of Fine Arts"]), df1.College), :]
clg_public_health=df1[findall(in(["College of Public Health"]), df1.College), :];
```

```julia
a=[clg_architecture,clg_humanities,clg_Agre_life_science,clg_applied_science,clg_pland_landscape,clg_education,clg_engineering,clg_optical,clg_medicine,clg_nurshing,clg_science,clg_social_bhvr,clg_eller,clg_fine_arts,clg_public_health];
```

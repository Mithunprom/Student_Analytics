```julia
#A1=sortrows(A, by= x=x[1])
A=convert(Matrix, df[:,[1,9,6,5,12]]);
A=sortslices(A,dims=1,by=x->x[2]+x[3]+x[4],rev=true);
A1=A[:,2:4];
A1=convert(Array{Float64,2}, A1);
A2=A[:,1];
A3=[174,126,120,120,120,120,120,120,120,120,120,120,120,120,120,120];
```

```julia
fnt=Plots.font("Aria",10);
aa=[3]
ab=A1[:,1]+A1[:,2]+A1[:,3]
a=ab+repeat(aa,15);
#round(Array{Float64,1},ab,[2])
ab=[round(x,digits=3) for x âˆˆ ab if x>0]
```

```julia
using PlotlyJS
plotlyjs()
B1=groupedbar(A1, bar_position = :stack, 
    bar_width=.7,
    xticks=(1:15,A2),
    xrotation=55,
    fillalpha=1,
    legend=:topright,
    labels=["Avg. Non-UA Units" "Avg. UA Units-Not Applied" "Avg. UA Units-Applied"],hover=A1)


#lo = Layout(;hovermode="closest")
#hover!(B1,  hovermode="closest")


[annotate!(i, a[i], text(ab[i],10), :color) for i in 1:15]
#[annotate!(i,40, text(A1[i,3],8), :color) for i in 1:15]
#[annotate!(i,A1[i,3]+A1[i,2]-3, text(A1[i,2],8), :color) for i in 1:15]
#[annotate!(i,A1[i,3]+A1[i,2]+A1[i,1]-3, text(A1[i,1],8), :color) for i in 1:15]

B2=plot!(.6:1:16, A3, linetype=:steppost,linestyle=:dot,labels="Minimum Required Credits",
    arrow = :arrow,
    linealpha = 0.5,
    linewidth = 4,
    linecolor = :red,hover=[174,174,126,120,120,120,120,120,120,120,120,120,120,120,120,120,120];)

ylims!((0,200))
#fnt=Plots.font("Aria",.001)
default(titlefont=fnt, guidefont=fnt, tickfont=fnt, legendfont=fnt)
plot!(size=(800,900))
display("text/html",B1)
```
